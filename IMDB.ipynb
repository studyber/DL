```python
import numpy as np
import pandas as pd
from tensorflow.keras.datasets import imdb
from tensorflow.keras.layers import *
from tensorflow.keras.models import Sequential, load_model
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score
```

```python
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=10000)
```

```python
x_train[0]
```

```python
# "vectorize"converts textual data (sequences) into a numerical representation 
def vectorize_sequences(sequences, dimension=10000):
    result = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        result[i, sequence] = 1     # Sets specific indices of results[i] to 1s
    return result

#convert the train_data into a matrix representation and assigns it to x_train.
x_train = vectorize_sequences(x_train)
#convert the test_data into a matrix representation and assigns it to x_test.
x_test = vectorize_sequences(x_test)
```

```python
x_train[0]
```

```python
print(y_train[0], y_train.dtype)
```

```python
#converting y into array of datatype of float
y_train = np.asarray(y_train).astype('float32')
y_test = np.asarray(y_test).astype('float32')
```

```python
print(y_train[0], y_train.dtype)
```

```python
#building the model
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(10000,)))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#training the model
model.fit(x_train, y_train, epochs=20, batch_size=512, validation_data=(x_test, y_test))
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print('Test loss: ', loss)
print('Test accuracy: ', accuracy)
```

```python
# Making Predictions for testing data
#small number like'1e-06' might be printed as 0.000001  
import sys
np.set_printoptions(suppress=True, threshold = sys.maxsize)

result = model.predict(x_test)
result
```

```python
y_pred = np.zeros(len(result))
for i,score in enumerate(result):
    if score > 0.5:
        y_pred[i] = 1
    else:
        0
```

```python
print("Positive review: 1 \nNegative review: 0\n")
print("Array of reviews: ", y_pred)
```

```python
MSE = mean_squared_error(y_test, y_pred)
print("Mean Squared Error: ", MSE)

MAE = mean_absolute_error(y_pred, y_test)
print("Mean Absolute Error: ", MAE)

Accuracy = accuracy_score(y_pred, y_test)
print("Accuracy score: ",Accuracy)
```
