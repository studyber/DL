```python
import pandas as pd
import numpy as np
from tensorflow.keras.layers import *
from tensorflow.keras.models import Sequential
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv('Boston-house-price-data.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df_x = df.drop(['MEDV'], axis=1)
df_y = df['MEDV']
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=1)
```

```python
reg = LinearRegression()
reg.fit(x_train, y_train)
lin_y_pred = reg.predict(x_test)
```

```python
mse = mean_squared_error(y_test, lin_y_pred)
mae = mean_absolute_error(y_test, lin_y_pred)

print("MSE: ", mse)
print("Accuracy", 100-mse)
print("MAE", mae)
```

```python
model = Sequential()
model.add(Dense(128, activation='relu', input_shape=(13,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(loss='mse', optimizer='adam', metrics=['mae'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, epochs=100, batch_size=4, validation_data=(x_test, y_test))
```

```python
lin_y_pred
```

```python
rnn_y_pred = model.predict(x_test)
print(rnn_y_pred)
```

```python
mser = mean_squared_error(y_test, rnn_y_pred)
maer = mean_absolute_error(y_test, rnn_y_pred)

print("MSE: ", mser)
print("Accuracy", 100-mser)
print("MAE", maer)
```
